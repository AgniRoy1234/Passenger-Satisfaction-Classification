# Chi Square Test

```python
from scipy.stats import chi2_contingency
```

```python
def chi_square(crosstab):
    stat, p, dof, expected = chi2_contingency(crosstab)
    print(p)
```

```python
for i in cat:
    print(i)
    ct=pd.crosstab(df1[i],df1["Satisfaction"])
    chi_square(ct)
```

As per the t tests,since all the p values are less than 0.05, we can conclude that there is significant dependency between the respective variables and the Satisfaction rating

# Decision Tree

```python
from sklearn import tree
```

```python
#Hyperparameter tuning using GridSearchCv
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
dt=DecisionTreeClassifier()

params={'criterion':['gini','entropy'],'max_depth':list(range(6,12)),'max_leaf_nodes':list(range(10,18)),
        'min_samples_leaf':list(range(10,14))}
clf = GridSearchCV(dt, params)
clf.fit(X_train,y_train)
clf.best_params_
```
